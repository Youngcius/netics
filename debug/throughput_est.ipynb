```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_logicals = np.arange(10, 101)
distances = np.arange(9, 50, 2)
time_meas = 10
throughputs = distances ** 3
```

```python
plt.figure(figsize=(8, 4))
plt.style.use('seaborn')
plt.plot(distances, throughputs, 'o')
plt.xlabel('Code distances')
plt.ylabel('Throughput (Gbps)')
plt.title('Throughput vs. distance (single logical qubit)', fontsize=15)
```
